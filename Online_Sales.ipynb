# PROCESO DE UNIÓN DE LOS DOS DATA FRAME

El dataframe df_Details se une al dataframe df_0rders en base a la columna 'Order ID'. El data frame construido **'Union'** contiene 1500 rows × 11 columns.

Se agregaron las columnas'Order Date', 'CustomerName', 'State', y 'City'.



```python
# Crear un nuevo dataframe llamado 'Union'
Union = pd.merge(df_Details, df_Orders, on='Order ID')
```

# CAMBIO DE TIPO DE DATO EN EL NUEVO DATA FRAME

En el nuevo data frame 'Union' aplicamos un cambio de tipo de dato en la columna 'Order Date' de texto a fecha.
Con el método .dtypes verificamos que se realizó exitosamente.

```python
#Pasar Order Date a fecha
Union['Order Date'] = pd.to_datetime(Union['Order Date'], errors='coerce', format='%d-%m-%Y')
```

```python
Union.dtypes
```

## A continuación verificamos el rango de las fechas del nuevo data frame: **las fechas van del 1 de enero de 2018 a 31 de diciembre de 2018**.

```python
min_date = Union['Order Date'].min()  # Fecha mínima
max_date = Union['Order Date'].max()  # Fecha máxima

print(f"Rango de fechas:\nDesde: {min_date}\nHasta: {max_date}")
```

### Presenta la distribución de las ventas en India según ciudades: Indore, Mumbai y Pune son las ciudades de mayores ventas.

```python
# Crear un Grafico de Barras en base a las columnas 'City' vs 'Amount'
Union.groupby('City')['Amount'].sum().plot(kind='bar', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

# GRÁFICO DE LÍNEAS

```python
# Creamos time_series como una copia completa del data frame 'Union'
time_series = Union.copy()

# Cambiamos el formato de la fecha: creamos una columna 'Mes' que contiene el nombre del mes correspondiente a la fecha en la columna 'Order Date'
time_series['Mes'] = time_series['Order Date'].dt.strftime('%B')

# Renombramos las categorías
renombre_categoria = {
    'Clothing': 'Ropa',
    'Electronics': 'Electrónica',
    'Furniture': 'Muebles'
}

# Reemplazamos los nombres de las categorías por sus equivalentes en español
time_series['Category'] = time_series['Category'].replace(renombre_categoria)

# Creamos el gráfico de líneas
plt.figure(figsize=(14, 7))
sns.lineplot(data=time_series, x='Mes', y='Amount', hue='Category', marker='o')

# Personalizamos el gráfico
plt.title('Evolución de Compras por Categoría a lo Largo del Año 2018')
plt.xlabel('Mes')
plt.ylabel('Monto de las Compras')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

pib.png

En la gráfica se observa el comportamiento de las compras en la India en 2018 según categoría de productos.
Llama la atención la caida del consumo en ropa y muebles en el mes de julio, que puede explicarse por la caida en el Producto Interno Bruto (PIB) del país ese mismo año, especialmente del segundo trimestre.
Esta caida de ventas no se verifica en la categoría de electrónicos debido a que ropa y muebles hasta cierto punto se consideran bienes de lujo, también porque los productos electrónicos son indispensables para los procesos de producción.

# HISTOGRAMA

```python

# Gráfico 1, histograma de 'Amount' con límite de eje x hasta 3000
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(Union['Amount'], bins=20, kde=False, color='skyblue')
plt.xlim(0, 3000)
plt.title('Amount')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico 2, histograma de 'Profit' con límite de eje x de -1000 a 1000
plt.subplot(1, 2, 2)
sns.histplot(Union['Profit'], bins=20, kde=False, color='salmon')
plt.xlim(-1000, 1000)
plt.title('Distribución de Profit')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Línea para distribución normal profit
mean_profit = Union['Profit'].mean()
std_profit = Union['Profit'].std()
x_profit = np.linspace(-1000, 1000, 100)

# Import the norm object from scipy.stats
from scipy.stats import norm

plt.plot(x_profit, norm.pdf(x_profit, mean_profit, std_profit) * len(Union) * (2000 / 20), color='red', linestyle='--')
#plot
plt.tight_layout()
plt.show()
```

```python
# Generar un Grafico de Histograma en base a la columna 'Quantity': Cantidad

from matplotlib import pyplot as plt
Union['Quantity'].plot(kind='hist', bins=20, title='Quantity')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

# GRÁFICO DE TORTA

```python
# renombrar categoría
category_counts = Union['Category'].replace({'Clothing': 'Ropa', 'Electronics': 'Electrónica', 'Furniture': 'Muebles'}).value_counts()

# Crear un gráfico de torta con porcentajes
plt.figure(figsize=(8, 6))
plt.pie(
    category_counts,
    labels=category_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('Dark2'),
    textprops={'fontsize': 12, 'fontweight': 'bold'}  # Aumenta el tamaño y pone en negrilla
)
plt.title('Distribución de Categorías en Porcentajes', fontsize=14, fontweight='bold')  # Título en negrilla y más grande
plt.axis('equal')
plt.show()
```
