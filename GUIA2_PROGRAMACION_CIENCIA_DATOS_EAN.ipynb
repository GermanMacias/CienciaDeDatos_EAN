```python
#Cargue de datos

df = pd.read_csv('/content/10.000_Empresas_mas_Grandes_del_Pa_s_20240822.csv')
df.head()
```

```python


# Función para eliminar tildes de un texto
def eliminar_tildes(text):
    if isinstance(text, str):
        text = unicodedata.normalize('NFKD', text)
        text = ''.join([c for c in text if not unicodedata.combining(c)])
    return text


# Limpieza de un carácter especial (ejemplo con '$')
caracter_a_limpiar = '$'
df.replace(caracter_a_limpiar, '', regex=True, inplace=True)

# reducir NITs de longitud > 9
df['NIT'] = df['NIT'].astype(str)
df['NIT'] = df['NIT'].apply(lambda x: x[:9] if len(x) > 9 else x)


df = df.applymap(lambda s: eliminar_tildes(s.lower()) if isinstance(s, str) else s) # Convertir todos los textos a minúsculas y eliminar tildes

# Modificar la columna 'CIUDAD DOMICILIO' para que solo quede la primera palabra antes de un espacio o guion
df['CIUDAD DOMICILIO'] = df['CIUDAD DOMICILIO'].apply(lambda x: x.split(' ')[0].split('-')[0])

# Ajuste de los tipos de datos: todas las columnas se mantienen como texto
df['CIUDAD DOMICILIO'] = df['CIUDAD DOMICILIO'].apply(lambda x: x.split(' ')[0].split('-')[0])

# Ajuste de los tipos de datos: todas las columnas se mantienen como texto
df['NIT'] = df['NIT'].astype('str')
df['RAZÓN SOCIAL'] = df['RAZÓN SOCIAL'].astype('str')
df['SUPERVISOR'] = df['SUPERVISOR'].astype('str')
df['REGIÓN'] = df['REGIÓN'].astype('str')
df['DEPARTAMENTO DOMICILIO'] = df['DEPARTAMENTO DOMICILIO'].astype('str')
df['CIUDAD DOMICILIO'] = df['CIUDAD DOMICILIO'].astype('str')
df['CIIU'] = df['CIIU'].astype('str')
df['MACROSECTOR'] = df['MACROSECTOR'].astype('str')
df['INGRESOS OPERACIONALES'] = df['INGRESOS OPERACIONALES'].astype('str')
df['GANANCIA (PÉRDIDA)'] = df['GANANCIA (PÉRDIDA)'].astype('str')
df['TOTAL ACTIVOS'] = df['TOTAL ACTIVOS'].astype('str')
df['TOTAL PASIVOS'] = df['TOTAL PASIVOS'].astype('str')
df['TOTAL PATRIMONIO'] = df['TOTAL PATRIMONIO'].astype('str')
df['Año de Corte'] = df['Año de Corte'].astype('str')  # También lo mantenemos como texto

# Mostrar los primeros registros para verificar
df.head()
```
