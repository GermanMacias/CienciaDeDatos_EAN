Tranformaciones tipo de dato y respectiva limpieza

```python
# Librerías
import pandas as pd
import unicodedata
import re

# Función para eliminar tildes de un texto
def eliminar_tildes(text):
    if isinstance(text, str):
        text = unicodedata.normalize('NFKD', text)
        text = ''.join([c for c in text if not unicodedata.combining(c)])
    return text

# Función para limpiar caracteres específicos en las columnas de interés
def limpiar_columna(text):
    if isinstance(text, str):
        text = text.replace('$', '')  # Eliminar el símbolo $
        text = text.replace('.', ',')  # Reemplazar punto por coma  ## aclaración por alguna razón no me dejaba trabajar con el punto
    return text

# Cargar el archivo CSV
df = pd.read_csv('/content/10.000_Empresas_mas_Grandes_del_Pa_s_20240822.csv')

# Aplicar la limpieza específica a las columnas monetarias
columnas_mod = ['INGRESOS OPERACIONALES', 'GANANCIA (PÉRDIDA)', 'TOTAL ACTIVOS', 'TOTAL PASIVOS','TOTAL PATRIMONIO']
for columna in columnas_mod:
    df[columna] = df[columna].apply(limpiar_columna)

# Reducción de los NITs de longitud mayor a 9
df['NIT'] = df['NIT'].astype(str)  # Mantener NIT como texto
df['NIT'] = df['NIT'].apply(lambda x: x[:9] if len(x) > 9 else x)

# Convertir todos los textos a minúsculas y eliminar tildes
df = df.applymap(lambda s: eliminar_tildes(s.lower()) if isinstance(s, str) else s)

# Modificar la columna 'CIUDAD DOMICILIO' para elimine después del "-"
df['CIUDAD DOMICILIO'] = df['CIUDAD DOMICILIO'].apply(lambda x: x.split(' ')[0].split('-')[0])
#Modificar la columna 'REGIÓN' para que elimine lo que hay antes del "-"
df['REGIÓN'] = df['REGIÓN'].apply(lambda x: x.split('-')[1].strip() if '-' in x else x)

# Ajuste de los tipos de datos, todas las columnas se mantienen como texto
df['NIT'] = df['NIT'].astype('str')
df['RAZÓN SOCIAL'] = df['RAZÓN SOCIAL'].astype('str')
df['SUPERVISOR'] = df['SUPERVISOR'].astype('str')
df['REGIÓN'] = df['REGIÓN'].astype('str')
df['DEPARTAMENTO DOMICILIO'] = df['DEPARTAMENTO DOMICILIO'].astype('str')
df['CIUDAD DOMICILIO'] = df['CIUDAD DOMICILIO'].astype('str')
df['CIIU'] = df['CIIU'].astype('str')
df['MACROSECTOR'] = df['MACROSECTOR'].astype('str')
df['INGRESOS OPERACIONALES'] = df['INGRESOS OPERACIONALES'].astype('str')
df['GANANCIA (PÉRDIDA)'] = df['GANANCIA (PÉRDIDA)'].astype('str')
df['TOTAL ACTIVOS'] = df['TOTAL ACTIVOS'].astype('str')
df['TOTAL PASIVOS'] = df['TOTAL PASIVOS'].astype('str')
df['TOTAL PATRIMONIO'] = df['TOTAL PATRIMONIO'].astype('str')
df['Año de Corte'] = df['Año de Corte'].astype('str')

# primeros datos tabla
df.head()
#### NOTA: dejé todo en string para facilidad de la actividad 2
```

```python
#Cambiar "," por "." que se había cambiado antes por erroes que generaba pero para pasar a numeric se hace el cambio nuevamente
columnas_mod = ['INGRESOS OPERACIONALES', 'GANANCIA (PÉRDIDA)', 'TOTAL ACTIVOS', 'TOTAL PASIVOS', 'TOTAL PATRIMONIO']
for columna in columnas_mod:
    df[columna] = pd.to_numeric(df[columna].str.replace(',', '.'), errors='coerce')



#Tabla
df
```
