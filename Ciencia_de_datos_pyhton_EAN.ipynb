```python
# Librerias

import pandas as pd

# Carga de datos
data = pd.read_csv('/content/census_income.csv').drop(columns=['Unnamed: 0'], errors='ignore') # se omite la fila 0 llamada 'Unnamed: 0'


data.insert(0, 'Personas', range(1, 1 + len(data))) # se agrega la fila 'Personas' al inicio del DataFrame remplazando la que estaba como 'Unnamed: 0'
```

```python


# Diccionario de las descripciones de las variables en base a https://archive.ics.uci.edu/dataset/20/census+income

descriptions = {
    'Personas': 'ID de la persona.',
    'age': 'Edad de la persona en años.',
    'workclass': 'Clase de trabajo de la persona (e.g., privado, gobierno).',
    'fnlwgt': 'Peso final utilizado en el análisis demográfico.',
    'education': 'Nivel educativo alcanzado por la persona.',
    'education-num': 'Número de años de educación completados.',
    'marital-status': 'Estado civil de la persona.',
    'occupation': 'Ocupación laboral de la persona.',
    'relationship': 'Relación familiar dentro del hogar.',
    'race': 'Raza de la persona.',
    'sex': 'Sexo de la persona.',
    'capital-gain': 'Ganancia de capital obtenida.',
    'capital-loss': 'Pérdida de capital sufrida.',
    'hours-per-week': 'Número de horas trabajadas por semana.',
    'native-country': 'País de origen de la persona.',
    'income': 'Nivel de ingresos (>50K o <=50K).'
}

# Tabla descriptiva: Cantidad de valores, Valores Distintos, Valores Nulos, Tipo de datos, Descripción

variables_info = pd.DataFrame({
    'Variable': data.columns,
    'Total de valores': data.count(),
    'Valores Distintos': data.nunique(),
    'Valores Nulos': data.isnull().sum(),
    'Tipo de dato': data.dtypes,
    'Característica': [descriptions.get(var, 'No especificada') for var in data.columns]
})

# 'imprimir' tabla
variables_info.reset_index(drop=True, inplace=True)
display(variables_info)


```

```python
# Descripción de variables, Que dato contiene y su conteo
variables_categoricas = data.select_dtypes(include=['object']).columns

for var in variables_categoricas:
    print(f"Variable: {var}")
    print(f"Frecuencia de los valores más comunes:\n{data[var].value_counts().head()}")
    print("\n")
```

### Analísis

Como resultado de la actividad 1, se agrupa las variables en una tabla a fin de mejorar su visualización sobre las variables como el tipo de dato y descripción, mostrando un diccionario de datos para su fácil manejo o al momento de usar las columnas necesarias para hacer una respectiva profundización cuando se requiera usar el archivo csv

A modo de tener una perspectiva mejor de los datos de hace un conteo ('count') de cuantas veces se repite los datos dentro de una misma columna para sabe que contiene las columnas y su proporción.
